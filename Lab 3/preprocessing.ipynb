```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/players_22.csv')
df
```

```python
df.columns
```

```python
all_cols = df.columns.tolist()
cols_to_remove = [col for col in all_cols if 'id' in col or 'url' in col]
df = df[[col for col in all_cols if col not in cols_to_remove]]
```

```python
df = df.drop(columns=['long_name', 'dob', 'body_type', 'real_face', 'player_tags', 'player_traits', 'work_rate'])
```

```python
pd.set_option('display.max_rows', None) 
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['value_eur', 'wage_eur', 'club_name', 'league_name', 'league_level', 'club_position', 'club_jersey_number'])
```

```python
df = df.drop(columns=['club_loaned_from', 'club_joined', 'nation_position', 'nation_jersey_number', 'release_clause_eur', 'goalkeeping_speed'])
```

```python
df[['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']] = df[['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']].fillna(0)
```
